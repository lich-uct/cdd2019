#### Otazka: pokud zredukujeme kazdou molekulu v DrugBank na scaffoldy, ktery nam vyjde jako nejcastejsi?

zaludnost: nemuzeme porovnavat primo scaffoldy tak, jak jsou ulozene, musime je prekonvertovat na napr. SMILES

```python
scaffolds2count = {}
for mol in drugbank:
    Chem.rdmolops.RemoveStereochemistry(mol)
    scaffold = MurckoScaffold.GetScaffoldForMol(mol)
    smiles = Chem.MolToSmiles(scaffold)
    try:
        scaffolds2count[smiles] += 1
    except KeyError:
        scaffolds2count[smiles] = 1
```

```python
display(scaffolds2count)
```

```python
scaffolds2count.items()
```

```python
scaffold_smiles = sorted(scaffolds2count.items(), key = lambda x: x[1], reverse=True)
```

```python
scaffold_smiles[:10]
```

```python
Draw.MolsToGridImage([Chem.MolFromSmiles(smi) for smi, count in scaffold_smiles[:10]], legends=[str(count) for smi, count in scaffold_smiles[:10]])
```

#### Nyni mame obdobny ukol, ale chceme se divat na frameworky

vyfiltrujeme si latky, pro ktere by toto mohlo zlobit (coz jsou anorganicky podivnosti a latky s komplexama zeleza, kobaltu, ruthenia, ..)

```python
carbon = Chem.MolFromSmarts('[#6]')
redundant_atoms = Chem.MolFromSmarts('[#26,#27,#44]')
def is_redundant(mol):
    if not mol.HasSubstructMatch(carbon):
        return True
    return mol.HasSubstructMatch(redundant_atoms)
```

```python
drugbank = [m for m in suppl if m and not is_redundant(m)]
```

```python
frameworks2count = {}
for i, mol in enumerate(drugbank):
    Chem.rdmolops.RemoveStereochemistry(mol)
    scaffold = MurckoScaffold.GetScaffoldForMol(mol)
    scaffold2 = MurckoScaffold.MakeScaffoldGeneric(scaffold)
    scaffold3 = MurckoScaffold.GetScaffoldForMol(scaffold2)
    smiles = Chem.MolToSmiles(scaffold3)

    try:
        frameworks2count[smiles] += 1
    except KeyError:
        frameworks2count[smiles] = 1
```

#### Dalsi ukol: udelame knihovnu, pro kazdy framework vybereme jednu molekulu

```python
framework_library = {}
for mol in drugbank:
    Chem.rdmolops.RemoveStereochemistry(mol)
    scaffold = MurckoScaffold.GetScaffoldForMol(mol)
    scaffold2 = MurckoScaffold.MakeScaffoldGeneric(scaffold)
    scaffold3 = MurckoScaffold.GetScaffoldForMol(scaffold2)
    smiles = Chem.MolToSmiles(scaffold3)

    try:
        framework_library[smiles].append(mol)
    except KeyError:
        framework_library[smiles] = [mol]
```

```python
import random
```

```python
picked = [random.choice(mols) for scaffold, mols in framework_library.items()]
len(picked)
```

```python
Draw.MolsToGridImage(picked[:10])
```
