#### vyplneni tabulky molecules molekulami z drugbank:

```python
def molecule2insertstring(mol, columns):
    values = []
    for c in columns:
        try:
            # vkladani dat do tabulky: hodnoty oddelene '' 
            #jelikoz jsou ' ve vzorcich, tak se musi zdvojit (jako escape character)
            value = "'{}'".format(mol.GetProp(c).replace("'","''")) 
        except KeyError:
            value = 'null'
        values.append(value)
    return "({})".format(", ".join(values))
```

```python
molecule2insertstring(drugbank[666],molecule_columns)
```

```python
#priprava prikazu
query = "insert into drugbank.molecule ({}) values {}".format(
    ", ".join(molecule_columns),                                                          
    ", ".join((molecule2insertstring(mol, molecule_columns) for mol in drugbank)))
```

```python
#query
```

```python
c.rollback()
cursor.execute(query)
```

```python
c.rollback()
cursor.execute('select id, DRUGBANK_ID from drugbank.molecule')
drugbank2databaseid = {drugbank: idx for idx, drugbank in cursor.fetchall()}
drugbank2databaseid
```

```python
def molecule2synonyms(m):
    try:
        synonyms = [syn.strip().replace("'","''") for syn in m.GetProp('SYNONYMS').split(";")]
        
    except KeyError:
        synonyms = []
        pass
    return synonyms

molecule2synonyms(drugbank[666])
```

```python
def molecule2insertstring(m, drugbank2database):
    databaseid = drugbank2database[m.GetProp("DRUGBANK_ID")]
    return ",".join(["({}, '{}')".format(databaseid, synonym) for synonym in molecule2synonyms(m)])
```

```python
molecule2insertstring(drugbank[666],drugbank2databaseid)
```

```python
query = "insert into drugbank.synonym (molid, name) values {}".format(','.join([
    molecule2insertstring(m,drugbank2databaseid) for m in drugbank 
    if molecule2insertstring(m,drugbank2databaseid)])) #aby se tam nevkladal prazdny retezec -> nebude tam 2x ,
query
```

```python
c.rollback()
cursor.execute(query)
```

```python
c.commit()
```
