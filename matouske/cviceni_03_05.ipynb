### Ukol: latky odpovidajici Lipinskeho pravidlum
- 5 pravidel
- kazda latka muze nesplnit jedno

```python
# LIPINSKI:

# 1. molekulova hmotnost <= 500
def lipinski_mw(m):
    return Descriptors.ExactMolWt(m) <= 500

# logP (rozdelovaci koeficient) < 5
def lipinski_logp(m):
    return Descriptors.MolLogP(m) < 5

# hydrogen bonds donors <= 5
def lipinski_hbd(m):
    return Descriptors.NumHDonors(m) <= 5

# hydrogen bonds acceptors <= 10
def lipinski_hba(m):
    return Descriptors.NumHAcceptors(m) <= 10

# false kdyz to nesplnuje 2 a vice pravidel (2 nebo vice funkci vrati false)
def lipinski(m):
    return sum((lipinski_mw(m), lipinski_logp(m), lipinski_hbd(m), lipinski_hba(m))) > 2
```

```python
print(lipinski_mw(mols[0]), Descriptors.ExactMolWt(mols[0]))
print(lipinski_logp(mols[0]), Descriptors.MolLogP(mols[0]))
print(lipinski_hbd(mols[0]), Descriptors.NumHDonors(mols[0]))
print(lipinski_hba(mols[0]), Descriptors.NumHAcceptors(mols[0]))
print(lipinski(mols[0]))
```

```python
Draw.MolToImage(mols[0], size=(1000,1000), fitImage=True)
```

```python
# ta, ktera nesplnuji
failed = [m for m in mols if not lipinski(m)]
len(failed)
```

```python
# ta, ktera splnuji
passed = [m for m in mols if lipinski(m)]
len(passed)
```

```python
len(failed) + len(passed)
```

### odfiltrovat struktury, ktere nesplnuji PAINS pravidla
- nacist smarts aprovest filtrovani
- -> nacist smarts a vyfiltrovat molekuly z drugbank

```python
# ulozeni pains:
pains = []
with open("p_l15.txt","r") as f:
    for line in f:
        smarts, idx = line.strip().split()
        pain = Chem.MolFromSmarts(smarts)
        pain.SetProp("pain_name", idx)
        pains.append(pain)
pains        
```

```python
pains[0].GetProp("pain_name")
```

```python
Draw.MolToImage(pains[0], size=(500,500), fitImage=True)
```

```python
# aplikuje molekulu na list patternu, vrati list delky patternu + true/false, zda tam je
def apply_substructures(m, substructures):
    return [m.HasSubstructMatch(s) for s in substructures]
```

```python
pains_hits = [apply_substructures(m,pains) for m in mols]
```

```python
print(pains_hits[0])
```

```python
pained_substances = [m for m, hits in zip(mols, pains_hits) if sum(hits)] # zip zarovna 2 listy nad sebe -> tupple
pained_substances
```

```python
Draw.MolsToGridImage(pained_substances, 
                     molsPerRow=2, 
                     subImgSize=(500, 500), 
                     legends=[m.GetProp("GENERIC_NAME") for m in pained_substances])
```
