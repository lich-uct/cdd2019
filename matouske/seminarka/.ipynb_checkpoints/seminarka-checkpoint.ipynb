### 2.3 MDS

Standardizace molekul:

```python
isOrganic = Chem.MolFromSmarts("[#6]")
hasInorganic = Chem.MolFromSmarts("[!#6;!#7;!#8;!#16;!Cl;!F;!Br;!I]")

_s = molvs.Standardizer()

def standardization(mol):
    mol = _s.standardize(mol)
    mol = _s.fragment_parent(mol) 
    
    if mol.HasSubstructMatch(isOrganic) and not mol.HasSubstructMatch(hasInorganic):
        mol = _s.charge_parent(mol)
        mol = _s.stereo_parent(mol)
        return mol
    else:
        return None
```

```python
clean_mols = [standardization(mol) for mol in ligands["Molecule_rdkit"]]
```

Overeni, ze standardizace probehla u vsech molekul - ulozim si jejich index:

```python
wrong = [i for i,mol in enumerate(clean_mols) if not mol]
```

```python
len(clean_mols), len(wrong)
```

U jedne molekuly se to nezdarilo:

```python
wrong = wrong[0]
wrong
```

```python
type(clean_mols[wrong])
```

```python
ligands["Canonical Smiles"][wrong]
```

```python
ligands["Molecule_rdkit"][wrong]
```

Molekula obsahuje deuterium, ale to by nemelo vadit, tak jsem ji standardizovala "rucne":

```python
t = _s.standardize(ligands["Molecule_rdkit"][wrong])
t = _s.fragment_parent(t)
t = _s.charge_parent(t)
t = _s.stereo_parent(t)
t
```

```python
clean_mols[wrong] = t
```

Napocitani Morganovych fingerprintu:

```python
fps =  [Chem.GetMorganFingerprintAsBitVect(m, 2, nBits=1024) for m in clean_mols]
len(fps)
```

```python
ligands["FP"] = fps
```

```python
ligands[:5]
```

Vypocet MDS a ulozeni dvodimenzionarnich dat:

```python
sparse_mat = distance.pdist(fps, metric="jaccard")
dist_mat = distance.squareform(sparse_mat)

clf = manifold.MDS(n_components=2, n_init=1, max_iter=100, dissimilarity="precomputed")
X_mds = clf.fit_transform(dist_mat)
```

```python
mds_frame = pd.DataFrame(X_mds, columns=["x","y"])
```

#### Zobrazeni vysledku MDS

```python
seaborn.scatterplot(x="x",y="y", data=mds_frame)
```

#### Vizualizace vysledku MDS s rozlisenym typem - EC50 nebo IC50

```python
mds_frame["typ"] = ligands["Standard Type"]
```

```python
seaborn.scatterplot(x="x",y="y", data=mds_frame, hue="typ")
```

## 3. Predikce aktivity a klasifikace latek

Rozdeleni molekul do testovaci a trenovaci sady:

```python
train, test = model_selection.train_test_split(ligands)
len(train), len(test)
```

### 3.1 Predikce aktivity

Nejprve jsem cross-validaci urcila, jaky je nejlepsi pocet stromu:

```python
param_grid = {"n_estimators":[10,20,30,40,50]}

cvalidation = model_selection.GridSearchCV(RandomForestRegressor(), 
                                           param_grid=param_grid) 
```

```python
cvalidation.fit(list(train["FP"]), list(train["pChEMBL Value"]))
```

```python
cvalidation.best_params_
```

Natrenovani regresniho modelu:

```python
predictor = RandomForestRegressor(n_estimators=cvalidation.best_params_['n_estimators'])
predictor.fit([fp for fp in train["FP"]], 
              list(train["pChEMBL Value"])) 
```

Predikce pChEMBL hodnoty testovaci sady:

```python
test["Predict Value"] = predictor.predict(list(test["FP"]))
```

#### Vykresleni predikovana vs. zmerena hodnota na testovacich datech

```python
seaborn.scatterplot(x = "pChEMBL Value",y = "Predict Value", data=test)
```

Predikce pChEMBL hodnoty trenovaci sady:

```python
train["Predict Value"] = predictor.predict(list(train["FP"]))
```

#### Vykresleni predikovana vs. zmerena hodnota na trenovacich datech

```python
seaborn.scatterplot(x = "pChEMBL Value",y = "Predict Value", data=train)
```

MSE testovacich a trenovacich dat:

```python
mse_test = sqrt(mean_squared_error(test["pChEMBL Value"],test["Predict Value"]))
mse_train = sqrt(mean_squared_error(train["pChEMBL Value"],train["Predict Value"]))
mse_test, mse_train
```

R<sup>2</sup> skore testovacich a trenovacich dat:

```python
r2_test = r2_score(test["pChEMBL Value"],test["Predict Value"])
r2_train = r2_score(train["pChEMBL Value"],train["Predict Value"])
r2_test, r2_train
```

### 3.2 Klasifikace

Urceni nejlepsiho poctu stromu cross-validaci:

```python
param_grid = {"n_estimators":[10,20,30,40,50]}

cvalidation_clas = model_selection.GridSearchCV(RandomForestClassifier(), 
                                           param_grid=param_grid) 
```

```python
cvalidation_clas.fit(list(train["FP"]), list(train["Standard Type"]))
```

```python
cvalidation_clas.best_params_
```

Natrenovani klasifikacniho modelu:

```python
classifier = RandomForestClassifier(n_estimators=cvalidation_clas.best_params_['n_estimators'])
classifier.fit([fp for fp in train["FP"]], list(train["Standard Type"]))
```

Klasifikace testovacich latek:

```python
test["Predict Type"] = classifier.predict(list(test["FP"]))
```

#### Matice zamen (confusion matrix) klasifikovanych latek testovaci sady

```python
test_conf_matrix = pd.crosstab(test["Standard Type"],test["Predict Type"],
                               rownames=["Standard Type"],colnames=["Predict Type"])
seaborn.heatmap(test_conf_matrix,annot=True)
```

Presnost (accuracy) klasifikace testovaci sady:

```python
accuracy_score(test["Standard Type"],test["Predict Type"])
```

Klasifikace trenovacich latek:

```python
train["Predict Type"] = classifier.predict(list(train["FP"]))
```

#### Matice zamen klasifikovanych latek trenovaci sady

```python
train_conf_matrix = pd.crosstab(train["Standard Type"],train["Predict Type"],
                               rownames=["Standard Type"],colnames=["Predict Type"])
seaborn.heatmap(train_conf_matrix,annot=True)
```

Presnost (accuracy) klasifikace trenovaci sady:

```python
accuracy_score(train["Standard Type"],train["Predict Type"])
```
