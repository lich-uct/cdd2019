```python
suppl = Chem.SDMolSupplier('../data/drugbank.sdf')
```

```python
zaznamy = 0
atomy = 0
mols = []
for x in suppl:
    zaznamy+=1
    try:
        mols.append(x)
        atomy += x.GetNumAtoms()
    except:
        continue
        
print("Záznamů celkem: ", zaznamy)
print("Atomů celkem: ", atomy)
```

```python
print("počet atomů: ", atomy)
print("počet záznamů: ", zaznamy)
print("delka seznamu mols: ", len(mols))
    
    
```

```python
# průměrný počet aromatických a nearomatických kruhů molekul v DrugBank:

mols[5]
```

```python
# priklad 1: benzen
#m = Chem.MolFromSmiles("c1ccccc1")

# priklad 2: cyklohexan
#m = Chem.MolFromSmiles("C1CCCCC1")

#priklad 3: molekula s více kruhy
#m = mols[5]
#rings = m.GetRingInfo().AtomRings() #tohle je uplně obyčejná ntice čísel -> indexy v kruhu

aromaticke_pocet = 0
nearomaticke_pocet = 0

for molecule in mols:
    if molecule == None:
        continue
    else:
        kruhy = molecule.GetRingInfo().AtomRings()
    for kruh in kruhy:
        a = molecule.GetAtomWithIdx(kruh[0])
        if a.GetIsAromatic():
            aromaticke_pocet = aromaticke_pocet + 1
        else:
            nearomaticke_pocet = nearomaticke_pocet + 1
print("průměrný počet AROMATICKÝCH kruhů na molekulu: ", aromaticke_pocet/zaznamy)
print("průměrný počet NEAROMATICKÝCH kruhů na molekulu: ", nearomaticke_pocet/zaznamy)
            

    


```

```python
with open("DrugBank_to_SMILESwINCHIkey", mode='w') as f:
    for i in mols:
        if i != None:
            print(Chem.MolToInchiKey(i), "\t", Chem.MolToSmiles(i), end="\n")
        else: 
            pass
```
